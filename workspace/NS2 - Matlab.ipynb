# Vehicles

```python
Vehicles = mat['Data'][0][0]['Vehicles']
print("Vehicles : ",Vehicles.shape)
print(Vehicles.dtype)
print(mat['Data'][0][0]['Vehicles'][0][0].dtype)

vehicles_serie = pd.DataFrame()
for v in Vehicles:
    #print(v[0].dtype)
    #print(v[0]['AxlePosition'][0][0][0])
    typ = v[0]['Type'][0][0][0]
    nAxle = v[0]['nAxle'][0][0][0]
    AxlePosition = v[0]['AxlePosition'][0][0][0]
    AxleWeights = v[0]['AxleWeights'][0][0][0]
    Velocity = v[0]['Velocity'][0][0][0]
    Kv = v[0]['Kv'][0][0][0]
    Cv = v[0]['Cv'][0][0][0]
    new_vehicle = {'type': [typ], 'nAxle': [nAxle],'AxlePosition':[AxlePosition],'AxleWeights':[AxleWeights],'Velocity':[Velocity],'Kv':[Kv],'Cv':[Cv]}
    vehicle_serie = pd.DataFrame(data=new_vehicle)
    frames = [vehicles_serie,vehicle_serie]
    vehicles_serie = pd.concat(frames)
vehicles_serie
```

# Fréquences

```python

FrequencyVehicles = mat['Data'][0][0]['FrequencyVehicles']
print("FrequencyVehicles : ",FrequencyVehicles.shape)
print(FrequencyVehicles[0][0])

freqs_serie = pd.DataFrame()
for f in FrequencyVehicles:
    #print(v[0].dtype)
    #print(v[0]['AxlePosition'][0][0][0])
    freq = f[0]
    new_freq = {'freq': [freq]}
    freq_serie = pd.DataFrame(data=new_freq)
    frames = [freqs_serie,freq_serie]
    freqs_serie = pd.concat(frames)
freqs_serie.head()
```
