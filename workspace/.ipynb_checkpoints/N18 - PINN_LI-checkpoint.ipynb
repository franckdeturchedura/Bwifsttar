```python
from sciann import math
import numpy as np
from scipy import signal
import tensorflow as tf

E = 35 #Module d'Young
I = 0.2 #Inertie (b*h³/12)
p = 2500 #masse volumique
S = 4.9    #surface
L = 35 #longueur de la poutre
TOL = 0.001


def delta_X(x,v,t):
  """
    Retourne delta X
  """

  X = round(x - v*t)
  return signal.unit_impulse(701,X)


"""L1 = p*S*math.diff(u,t,order=2) + E*I*math.diff(u,x,order=4) - delta_X() #equa diff


C2  = (1+math.sign(-x + TOL)) * (u)  #contrainte u(x=0)=0
C3  = (1-math.sign(x - (L-TOL))) * (u) #contrainte u(x=L)=0
C4  = (1+math.sign(-x + TOL)) * math.diff(u,x,order=2) #contrainte u''(x=0)=0
C5  = (1-math.sign(x - (L-TOL))) * math.diff(u,x,order=2) #contrainte u''(x=L)=0"""

def constraints(x,v,t):
  L1 = p*S*math.diff(u,t,order=2) + E*I*math.diff(u,x,order=4) - delta_X(x,v,t) #equa diff


  C2  = (1+math.sign(-x + TOL)) * (u)  #contrainte u(x=0)=0
  C3  = (1-math.sign(x - (L-TOL))) * (u) #contrainte u(x=L)=0
  C4  = (1+math.sign(-x + TOL)) * math.diff(u,x,order=2) #contrainte u''(x=0)=0
  C5  = (1-math.sign(x - (L-TOL))) * math.diff(u,x,order=2) #contrainte u''(x=L)=0

  return [L1,C2,C3,C4,C5]

def def_model(x,v,t):   
  model = sn.SciModel([x,t],constraints(x,v,t) , "mse", "Adam") #modèle basé sur Mean Squared Error + Adam
  return model
```

### II.1 Définition de l'équation différentielle à résoudre 

```python
 
E = 35 #Module d'Young
I = 0.2 #Inertie (b*h³/12)
p = 2500 #masse volumique
S = 4.9    #surface
L = 33 #longueur de la poutre
 

def pde(x,u):
    
    """"définit l'equa diff"""

    def delta_X(x):
        X = x[0]
        if tf.round(X)==0:
            d = tf.constant(1.0)
        else:
            d = tf.constant(0.0)
        return d

    du_x = tf.gradients(u,x)[0] # dérivées premières de u par rapport à x et t

    du_x, du_t = du_x[:, 0:1], du_x[:, 1:2] 

    du_xx = tf.gradients(du_x, x)[0][:, 0:1]
    du_tt = tf.gradients(du_x, x)[0][:, 1:2]

    du_xxx = tf.gradients(du_xx, x)[0][:, 0:1]

    du_xxxx = tf.gradients(du_xxx, x)[0][:, 0:1]

    #X = x[:,0:1] - 18.96* x[:,1:2]
    #print("X : ,",X)
    v=18.96
    return E*I*du_xxxx + p*S*du_tt - delta_X(x)
```

```python
def ddu(x, u):
    du_x = tf.gradients(u, x)[0]
    du_x = du_x[:, 0:1]
    return tf.gradients(du_x, x)[0][:, 0:1]
```

```python
v = 18.96
geom = dde.geometry.Interval(0, L)
timedomain = dde.geometry.TimeDomain(0, L/v)
geomtime = dde.geometry.GeometryXTime(geom, timedomain)
```

### II.2 Définition des conditions aux limites

```python
def boundary_r(x, on_boundary):
#return on_boundary and np.isclose(x[0],0)

    return np.isclose(x[0],0)
bc0 = dde.OperatorBC( geomtime, lambda x, u, _: u, boundary_r)    #u(t,x=0)=0


def boundary_l(x, on_boundary):
    return np.isclose(x[0], L)
bc1 = dde.OperatorBC(geomtime, lambda x, u, _: u, boundary_l)    #u(t,x=L)=0



def boundary_r1(x, on_boundary):
#return on_boundary and np.isclose(x[1],L/v)

    return np.isclose(x[1],L/v,rtol=0.001)
bc2 = dde.OperatorBC( geomtime, lambda x,u,_ :u , boundary_r1)   #u(t=inf,x) = 0



def boundary_l1(x, on_boundary):
#return on_boundary and np.isclose(x[1],0)
    return np.isclose(x[1],0,rtol=0.0001)
bc3 = dde.OperatorBC( geomtime, lambda x,u,_ :u , boundary_l1)   #u(t=0,x) = 0



def boundary_r2(x, on_boundary):
#return on_boundary and np.isclose(x[1],L/v)
    return np.isclose(x[0],0)
bc4 = dde.OperatorBC(geomtime, lambda x, u, _: ddu(x, u), boundary_r2)   #u''(t,x=0)=0




def boundary_l2(x, on_boundary):
#return on_boundary and np.isclose(x[1],0)
    return np.isclose(x[0],L)
bc5 = dde.OperatorBC( geomtime, lambda x, u, _: ddu(x, u), boundary_l2)   #u''(t,x=L)=0

```

### II.3 Définition de la solution de l'équa diff comme neural network

```python
data = dde.data.TimePDE( geomtime,pde,[bc0,bc1,bc2,bc3,bc4,bc5],num_domain=2546,num_boundary=160,num_initial=160)
 
net = dde.maps.FNN( [2] + [32]*2 +[64]+[32] +[1], "tanh", "Glorot normal")
```

### II.4 Résolution

```python

from deepxde import model

model = dde.Model(data,net)
model.compile("adam", lr=1e-3)
model.train(epochs=1000)
model.compile("L-BFGS-B")
losshistory, train_state = model.train()
```
