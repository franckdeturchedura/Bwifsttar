# Avec resampling de la LI

```python
from scipy.signal import resample

def LI_resampling(influence,freq):
    x = np.arange(0,len(influence))
    h1, x1 =resample(influence,freq,x)
    #plt.plot(x, influence, x1, h1)
    #plt.show()
    return h1,x1

#h1,x1 = LI_resampling(h,5000)
def time_to_meter_sampling_mt(truck, influence):
    func, dist = time_to_meter_interpolation_mt(truck, influence)
    meters = truck.speed * truck.time
    meters = meters[meters<=dist]
    return func(meters)#donne les valeurs via fonction d'interpolation des 'influence' (temps) pour une distance maximum

def time_to_meter_interpolation_mt(truck, influence): 
    #meters = truck.speed * truck.time[:len(influence)]#metres parcourus sur le temps len(influence)
    #meters = np.linspace(0,(truck.speed*truck.time).max(),10000)
    meters = np.linspace(0,30,10000)
    dist   = meters.max()#le nombre de metres total
    print(meters.shape)
    print(influence.shape)
    func2   = interp1d(meters, influence, fill_value="extrapolate")#permet à partir de meters et infuence de trouver une approximation de influence = f(meters)
    return func2, dist#retourne la distance totale et la fonction d'interpolation
```

```python
calibration_trucks,traffic_trucks = load_senlis_modified(selected=6) #[3,4,6,7]
trucks = [truck for truck in calibration_trucks if(len(truck.weights) == len(truck.peaks))]

list_speeds = []
list_infl = []

for i in range(len(trucks)):
    list_speeds.append(trucks[i].speed)
    list_infl.append(calibration_mt(trucks[i:i+1],l2_reg={'strength': 1e3, 'cutoff': 0.01},tv_reg={'strength': 1e3, 'cutoff': 0.95}))
list_infl = np.array(list_infl)
list_speeds = np.array(list_speeds)

print(list_speeds.shape)
print(list_infl.shape)
```

```python
def estimation(truck, influence_bundle):
    influence= time_to_meter_sampling_mt(truck, influence_bundle)
    toeplitz = create_toeplitz(truck.signals.shape[-1], influence.size, truck.peaks)
    H_matrix = toeplitz @ influence
    w ,_,_,_ = np.linalg.lstsq(H_matrix.T, truck.signals, rcond=None)
    return w
def estimation_mt(trucks, influence_bundle):
    list_w = []
    trucks = [truck for truck in trucks if(len(truck.weights) == len(truck.peaks))]
    
    for truck in trucks:
        
        influence= time_to_meter_sampling_mt(truck, influence_bundle)
        toeplitz = create_toeplitz(truck.signals.shape[-1], influence.size, truck.peaks)
        H_matrix = toeplitz @ influence
        w ,_,_,_ = np.linalg.lstsq(H_matrix.T, truck.signals, rcond=None)
        list_w.append(w)
    return list_w

```
