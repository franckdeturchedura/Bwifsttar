## II/ Signal théorique

## III/ Resampling via scipy.signal.resample

```python
fqe = 0.002
##rééchantilloner et regarder maxs
```

```python
from scipy.signal import resample
infl = df['E (Pa)']
meters = df['x (m)']
print(infl.shape)
print(meters.shape)
```

```python
infl_300 = resample(infl,300)
meters_300 = resample(meters,300)
print(infl_300.shape)
print(meters_300.shape)
```

```python
%matplotlib notebook
plt.figure()
plt.plot(df['x (m)'],df['E (Pa)'],label=" Fréquence échantillonage : 0.002 (base)")

for i in range(50,400,50):
    plt.plot(resample(meters,i),resample(infl,i),label=str(0.02/i))

plt.legend()
plt.show()
```

## IV/ Resampling par "oubli" de valeurs

```python
%matplotlib notebook
plt.figure()

for i in range(1,30,5):
    fq = np.arange(0,len(infl),i)
    infl_demi = infl[fq]
    meters_demi = meters[fq]
    print(infl_demi.shape)
    print(meters_demi.shape)
    plt.plot(meters_demi,infl_demi,label=str(0.002/i))
plt.plot(meters,infl,label="normale")
plt.legend()
plt.show()
```
