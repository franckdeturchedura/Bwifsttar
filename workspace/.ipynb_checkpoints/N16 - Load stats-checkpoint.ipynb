```python

plt.figure()
sigma = list_erreurs_tot.std()
mu = list_erreurs_tot.mean()
print("MU :",mu)
print("SIGMA : ",sigma)
b= np.linspace(-15,15,2000)
plt.plot(b, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (b - mu)**2 / (2 * sigma**2) ),'o',linewidth=0.5, color='r')
a = 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (list_erreurs_tot - mu)**2 / (2 * sigma**2) )
print("dims : ",a.shape)

print((list_erreurs_tot).shape)
#plt.figure()
sns.distplot(list_erreurs_tot, hist=True, rug=True);
plt.title("Erreurs sur les poids totaux redressés (t)")
plt.show()
```

```python
import scipy.stats
from scipy import stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a) #scipy.stats.sem : Compute standard error of the mean.

    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

print("Intervalle de confiance à 95% : ",stats.norm.interval(0.95, loc=mu, scale=sigma))
```
